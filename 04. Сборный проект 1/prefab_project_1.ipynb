# Сборный проект-1: "Выявление закономерностей определяющих успешность игр"

**Задание:**

**Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.**

**Описание проекта**

**Перед нами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли мы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.**

**Описание данных:**

* Name — название игры
* Platform — платформа
* Year_of_Release — год выпуска
* Genre — жанр игры
* NA_sales — продажи в Северной Америке (миллионы проданных копий)
* EU_sales — продажи в Европе (миллионы проданных копий)
* JP_sales — продажи в Японии (миллионы проданных копий)
* Other_sales — продажи в других странах (миллионы проданных копий)
* Critic_Score — оценка критиков (максимум 100)
* User_Score — оценка пользователей (максимум 10)
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.


**Примечание:**
**В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».**

*	Столбец critic_score (оценка критиков) — вполне возможно, что игры просто не оценивались критиками, из-за отсутствия механизма оценки или большого разброса оценки по разным ресурсам. Чтобы иметь возможность работать с данными в полном объёме заменим NaN на -1, можем это сделать т. к. "10" не встречается в critic_score, но при этом будем иметь в виду, что это "-1" - отсутствие критики.
*	Столбец user_score (оценка пользователей) — видимо аналогичная причина с critic_score, но здесь еще встречается значение tbd (To Be Determined), что означит что пока нет значения, поэтом у и NaN и tbd заменим на значение, которое не встречается в столбце и не находится рядом с другими значениями, например -1, но при этом будем иметь в виду, что это "-1" это отсутствие оценки пользователей.
*	Столбец rating (рейтинг от организации ESRB) — Эта организация Entertainment Software Rating Board была создана в 1994 году и видимо или сама ESRB не сделала рейтинги игр, появившихся до её создания или просто технически рейтинг ESRB, был частично утерян при копировании этой базы или просто не был предоставлен на этапе формирования базы, а может и то и другое вместе. Заменим NaN на "No_rating"


```python
print(games.shape)
games.drop(labels = [659, 14244],axis = 0, inplace = True)
games['year_of_release'] = games['year_of_release'].fillna(-1)
games['critic_score'] = games['critic_score'].fillna(-1)
games['user_score'] = games['user_score'].replace('tbd', np.NaN)
games['user_score'] = games['user_score'].fillna(-1)
games['rating'] = games['rating'].fillna('No_rating')
print(games.isna().sum())
table(games)
print(games.shape)
```

* Заменим user_score на float, т. к. это категориальные текстовые данные.

#### Напишем функцию выбора ТОП платформ с наибольшими суммарными продажами и построением распределения продаж по годам, а также общего суммарного распределения выбранного количества ТОП платформ. 

```python
def biggest_sales(top, df):
    
    games_grup = df.groupby( 'platform', as_index=False) \
                                       .aggregate({'sales_total': 'sum'}) \
                                       .sort_values(by='sales_total', ascending=False,).head(top)
    list_y = games_grup['platform'].values
    n = 0
    for i in list_y:
        n += 1
        if n == 1:
            ax = df[df['platform'] == i].groupby( 'year_of_release', as_index=False) \
                                       .aggregate({'sales_total': 'sum'}).plot(x='year_of_release'
                                             , y='sales_total'
                                             , kind='bar'
                                             , ec='black' 
                                             , fontsize=15
                                             , figsize=(18,5)
                                             , legend=None
                                             , color = 'g')

            i_text = 'Распределение по годам суммарных продаж по платформе ' + i
            plt.title(i_text, fontsize=20, fontweight="bold")
            plt.xlabel("Года", fontsize=14, fontweight="bold")
            plt.ylabel("Суммарные продажи, млн. копий", fontsize=14, fontweight="bold")
               
            games_grup = df[df['platform'] == i].groupby( 'year_of_release', as_index=False) \
                                       .aggregate({'sales_total': 'sum'})
            games_grup.rename(columns = {'sales_total': i}, inplace = True)
        elif n > 1:
            ax = df[df['platform'] == i].groupby( 'year_of_release', as_index=False) \
                                       .aggregate({'sales_total': 'sum'}).plot(x='year_of_release'
                                             , y='sales_total'
                                             , kind='bar'
                                             , ec='black' 
                                             , fontsize=15
                                             , figsize=(18,5)
                                             , legend=None
                                             , color = 'g')

            i_text = 'Распределение по годам суммарных продаж по платформе ' + i
            plt.title(i_text, fontsize=20, fontweight="bold")
            plt.xlabel("Года", fontsize=14, fontweight="bold")
            plt.ylabel("Суммарные продажи, млн. копий", fontsize=14, fontweight="bold")
            
            games_grup = pd.merge(games_grup
                                , df[df['platform'] == i].groupby( 'year_of_release', as_index=False) \
                                                               .aggregate({'sales_total': 'sum'})
                                , on='year_of_release'
                                , how='outer')
            games_grup.rename(columns = {'sales_total': i}, inplace = True)
                  
    games_grup = games_grup.loc[games_grup['year_of_release'] > 1950] 
        
    games_grup.plot(x='year_of_release', y=list_y, figsize=(18,10), kind='line', fontsize=10)
    plt.title('Распределение по годам суммарных продаж топовых платформ', fontsize=20, fontweight="bold")
    plt.xlabel("Года", fontsize=14, fontweight="bold")
    plt.ylabel("Суммарные продажи, млн. копий", fontsize=14, fontweight="bold")
    
    plt.show()  
```

Напишем функцию построения boxplot

```python
def boxplot_fun(ylims=None):
    games_grup_top = games_current_period.groupby( 'platform', as_index=False) \
                                     .aggregate({'sales_total': 'sum'}) \
                                     .sort_values(by='sales_total', ascending=False,).head(6)
                             
    top_platforms = games_grup_top['platform'].values
    query_condition="platform in @top_platforms"
    df_query = games_current_period.query(query_condition)
    by_column = query_condition.split()[0]
    ax = df_query.boxplot(
        column='sales_total',
        by=by_column, 
        fontsize=25, 
        figsize=(25,15),
        boxprops=dict(
            linestyle='-',
            linewidth=5
        ), medianprops=dict(
            linestyle='-', 
            linewidth=5
        ), whiskerprops=dict(
            linestyle='-', 
            linewidth=5
        ), capprops=dict(
            linestyle='-',
            linewidth=5
        ), notch=True
    )
    ax.set_xlabel('')
    ax.set_title('')
    if ylims != None:
        ax.set_ylim(ylims)
    plt.suptitle('Boxplot продаж по топовым платформам', y=0.94, fontsize=35, fontweight='bold')
    
```

```python
def describe_fun(top):
    games_grup_top = games_current_period.groupby( 'platform', as_index=False) \
                                     .aggregate({'sales_total': 'sum'}) \
                                     .sort_values(by='sales_total', ascending=False,).head(top)
                             
    top_platforms = games_grup_top['platform'].values
    n = 0
    for i in top_platforms:
        n += 1
        if n == 1:
            data_describe = games_current_period[games_current_period['platform'] == i].describe()[['sales_total']]
            data_describe.rename(columns = {'sales_total': i}, inplace = True)
            data_describe.reset_index(inplace = True)
            data_describe.rename(columns = {'index': 'condition'}, inplace = True)
        else:
            data = games_current_period[games_current_period['platform'] == i].describe()[['sales_total']]
            data.rename(columns = {'sales_total': i}, inplace = True)
            data.reset_index(inplace = True)
            data.rename(columns = {'index': 'condition'}, inplace = True)
            data_temp = pd.merge(data_describe, data, on='condition', how='right')
            data_describe = data_temp
            
    return display(data_describe.set_index('condition'))

    
    
```

```python
ESRB_dict = { 'T': 'Подросткам'             
             , 'No_rating': 'Нет рейтинга'
             , 'M': 'Для взрослых'                        
             , 'E10+': 'Для всех от 10 лет и старше'
             , 'E': 'Для всех'}
```

Добавим новый столбец в основной датафрейм с разъяснением

```python
#games_current_period['ESRB_description'] = games_current_period['rating'].map(ESRB_dict)# и так что-то ругается
def ESRB_rating(row):    
    rating = row['rating']
    if rating  == 'T':
        return 'Подросткам'
    elif rating == 'No_rating':
        return 'Нет рейтинга'
    elif rating == 'M':
        return 'Для взрослых'
    elif rating == 'E10+':
        return 'Для всех от 10 лет и старше'    
    elif rating == 'E':
        return 'Для всех' 
     
games_current_period['ESRB_description'] = games_current_period.apply(ESRB_rating, axis = 1)
```

### Проверка гипотез

#### **Гипотеза первая:** Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

* Н0 (нулевая гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
* Н1 (альтернативная гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой.

Критический уровень значимости поставим 0.05.

Воспользуюсь методом scipy.stats.ttest_ind(array1, array2, equal_var = False)

```python
xbox_one_ratings = games_current_period[(games_current_period['platform'] == 'XOne') 
                                        & (games_current_period['user_score'] > 0)]
pc_ratings = games_current_period[(games_current_period['platform'] == 'PC') 
                                  & (games_current_period['user_score'] > 0)]

display(xbox_one_ratings.pivot_table(index = 'platform', values = 'user_score', aggfunc = 'count'))
display(pc_ratings.pivot_table(index = 'platform', values = 'user_score', aggfunc = 'count'))

alpha = 0.05
results = st.ttest_ind(xbox_one_ratings['user_score'], pc_ratings['user_score'], equal_var = False)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

#### **Вывод:** Полученное нами p-значение равно 0.12, а значит есть вероятность получить такие или большие различия при условии, что средние на самом деле равны и мы не можем отвергнуть нулевую гипотезу, которая говорит о равенстве средних значений.

**Значит Н1 (альтернативная гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой, отвергается.**

#### **Гипотеза вторая:** Средние пользовательские рейтинги жанров Action  и Sports разные.

* Н0 (нулевая гипотеза) - средние пользовательские рейтинги жанров Action и Sports одинаковые.
* Н1 (альтернативная гипотеза) - средние пользовательские рейтинги жанров Action и Sports отличаются между собой.

Критический уровень значимости поставим 0.05.

Воспользуюсь методом scipy.stats.ttest_ind(array1, array2, equal_var = False)

```python
action_ratings = games_current_period[(games_current_period['genre'] == 'Action') 
                                      & (games_current_period['user_score'] > 0)]
pc_ratings = games_current_period[(games_current_period['genre'] == 'Sports') 
                                  & (games_current_period['user_score'] > 0)]

display(action_ratings.pivot_table(index = 'genre', values = 'user_score', aggfunc = 'count'))
display(pc_ratings.pivot_table(index = 'genre', values = 'user_score', aggfunc = 'count'))

alpha = 0.05
results = st.ttest_ind(action_ratings['user_score'], pc_ratings['user_score'], equal_var = False)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

####  **Вывод:** Нулевой гипотезой была гипотеза о равенстве средних  и очень маленькое p-значение говорит нам о том, вероятность случайно получить такие или более выраженные различия средних при равных средних мала. Нулевая гипотеза отвергается. 

**Значит Н1 (альтернативная гипотеза) - средние пользовательские рейтинги жанров Action и Sports отличаются между собой, не отвергается.**
