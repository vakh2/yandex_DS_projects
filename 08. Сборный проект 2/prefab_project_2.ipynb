# Стадии технологического процесса восстановления золота из руды

 **1. Флотация**
 
Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.

На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

**2. Очистка**

Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

# Описание данных
* Технологический процесс
 * Rougher feed — исходное сырье
 * Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
   * Xanthate **— ксантогенат (промотер, или активатор флотации);
   * Sulphate — сульфат (на данном производстве сульфид натрия);
   * Depressant — депрессант (силикат натрия).
 * Rougher process (англ. «грубый процесс») — флотация
 * Rougher tails — отвальные хвосты
 * Float banks — флотационная установка
 * Cleaner process — очистка
 * Rougher Au — черновой концентрат золота
 * Final Au — финальный концентрат золота
* Параметры этапов
 * air amount — объём воздуха
 * fluid levels — уровень жидкости
 * feed size — размер гранул сырья
 * feed rate — скорость подачи

# Наименование признаков
* Наименование признаков должно быть такое:
 * [этап].[тип_параметра].[название_параметра]
 * Пример: rougher.input.feed_ag
* Возможные значения для блока [этап]:
 * rougher — флотация
 * primary_cleaner — первичная очистка
 * secondary_cleaner — вторичная очистка
 * final — финальные характеристики
* Возможные значения для блока [тип_параметра]:
 * input — параметры сырья
 * output — параметры продукта
 * state — параметры, характеризующие текущее состояние этапа
 * calculation — расчётные характеристики

* F — доля золота в сырье/концентрате до флотации/очистки;
* T — доля золота в отвальных хвостах после флотации/очистки.

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.

Для решения задачи необходимо использовать метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).

### Посмотрим, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах очистки

Напишем функцию для сбора данных вместе

```python
def column_selection(df, metal):
    met = 'au'
    away = 'tail_'
    if metal == 'Золото':        
        met = 'au'
    elif metal == 'Серебро':
        met = 'ag'
    elif metal == 'Свинец':   
        met = 'pb'
    away = away + met   
    columns = list(set(df.columns[df.columns.str.endswith(met)].tolist())
             - set(df.columns[df.columns.str.endswith(away)].tolist()))
    columns.sort(reverse=True)
    columns[0],columns[1] = columns[1],columns[0]
    return columns
```

```python
columns_au = column_selection(df_full, 'Золото')
print(columns_au)
columns_ag = column_selection(df_full, 'Серебро')
print(columns_ag)
columns_pb = column_selection(df_full, 'Свинец')
print(columns_pb)
```

### Напишем функцию для вычисления sMAPE

Это нам нужно чтоб отслеживать предварительные данные на этапе после флотации

```python
def smape_fun(target, predict):
    smape = 100/len(target) * np.sum(2 * np.abs(target - predict) / 
                                                     (np.abs(target) + np.abs(predict)))
    return smape
```

### Напишем функцию для вычисления итоговой sMAPE

```python
def final_smape_fun(target_rougher, predict_rougher, target_final, predict_final):
    smape_rougher = 100/len(target_rougher) * np.sum(2 * np.abs(target_rougher - predict_rougher) / 
                                                     (np.abs(target_rougher) + np.abs(predict_rougher)))
    smape_final = 100/len(target_final) * np.sum(2 * np.abs(target_final - predict_final) / 
                                                 (np.abs(target_final) + np.abs(predict_final)))
    smape = 0.25*(smape_rougher) + 0.75*(smape_final)
    return smape
```

### Напишем функцию разделения данных c отображением количества строк.

Функция нужна для использования по-разному подготовленных данных.
Для этапа флотации в обучающей выборке оставим только относящиеся к флотации признаки.
Для этапа очистки оставим в обучающей выборке оставим только относящиеся к очистки признаки.

```python
def splitting(train, test):
    features_train_rougher = train.drop(['rougher.output.recovery'], axis=1).filter(regex='^rougher',axis=1)
    
    features_train_final = train.drop((features_train_rougher.columns), axis=1) \
                                .drop(['final.output.recovery', 'date'], axis=1)
    target_train_rougher = train['rougher.output.recovery']
    target_train_final = train['final.output.recovery']
    
    features_test_rougher = test.drop(['rougher.output.recovery'], axis=1).filter(regex='^rougher',axis=1)
    
    features_test_final = test.drop((features_test_rougher.columns), axis=1) \
                                .drop(['final.output.recovery', 'date'], axis=1)
    
    target_test_rougher = test['rougher.output.recovery']
    target_test_final = test['final.output.recovery']
    print('Размер обучающей выборки rougher:', features_train_rougher.shape, 'пропусков:'
          , features_train_rougher.isna().sum().sum())
    print('Размер обучающей выборки final:', features_train_final.shape, 'пропусков:'
          , features_train_final.isna().sum().sum())
    print('Размер target_train_rougher:', target_train_rougher.shape, 'пропусков:', target_train_rougher.isna().sum().sum())
    print('Размер target_train_final:', target_train_final.shape, 'пропусков:', target_train_final.isna().sum().sum())
    
    print('Размер тестовой выборки rougher:', features_test_rougher.shape
          , 'пропусков:', features_test_rougher.isna().sum().sum())   
    print('Размер тестовой выборки final:', features_test_final.shape
          , 'пропусков:', features_test_final.isna().sum().sum())        
    
    print('Размер target_test_rougher:', target_test_rougher.shape, 'пропусков:', target_test_rougher.isna().sum().sum())
    print('Размер target_test_final:', target_test_final.shape, 'пропусков:', target_test_final.isna().sum().sum())
    return (
        features_train_rougher,
        features_train_final,
        target_train_rougher,
        target_train_final,
        features_test_rougher,
        features_test_final,
        target_test_rougher,
        target_test_final
    )
```

### Обучим разные модели и оценим их качество кросс-валидацией.

#### Напишем функцию, которая перебором, меняя параметры ищет лучшую модель решающего дерева по минимальному значению итогового sMAPE в ходе перекрёстной проверки на обучающей выборке с cv=5

```python
def search_best_parameters_tree(features_train_rougher, features_train_final, target_train_rougher, target_train_final):    
  
    best_model_tree_rougher = None
    best_model_tree_final = None
    smape = 1000
    best_smape_rougher = 1000
    best_smape_final = 1000
    final_smape_all = 1000
    i = 0
    for access in [target_train_rougher, target_train_final]:
        i = i + 1
        if i == 1:
            features_train = features_train_rougher
        elif i == 2:
            features_train = features_train_final
        for depth in range(2,6):
            for crit in ['mse', 'mae']:
                for split in ["best", "random"]:
                    model = DecisionTreeRegressor(random_state=12345
                                                   , max_depth=depth
                                                   , criterion=crit
                                                   , splitter=split)
                                                  
                    model.fit(features_train, access)
                    predictions = model.predict(features_train)
                    predictions_cross = cross_val_predict(model, features_train, access, cv=5)
                    smape = smape_fun(access, predictions_cross)
                    
                    if i == 1:
                        if smape < best_smape_rougher:
                            best_split=split
                            best_crit=crit
                            max_depth=depth
                            predictions_cross_rougher = predictions_cross
                            best_smape_rougher = smape
                            best_model_tree_rougher = model
                        print(depth, crit, split, '|best_smape_rougher =', best_smape_rougher, '|smape = ', smape,)
                    elif i == 2:
                            smape_all = final_smape_fun(target_train_rougher
                                                              , predictions_cross_rougher
                                                              , access
                                                              , predictions_cross)
                            if smape_all < final_smape_all:
                                final_smape_all = smape_all    
                                best_split=split
                                best_crit=crit
                                max_depth=depth
                                best_smape_final = smape
                                best_model_tree_final = model

                            print(depth, crit, split, '|best_smape_final =', best_smape_final
                                      , '|smape = ', smape
                                      , '|final_smape_all = ', final_smape_all
                                      , '|smape_all = ', smape_all)
    return (best_model_tree_rougher
             , best_model_tree_final
             , best_smape_rougher
             , best_smape_final
             , final_smape_all)
```
